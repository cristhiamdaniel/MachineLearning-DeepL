# DATASET BOSTON (REGRESION)

```python
boston = load_boston()
print(boston.DESCR)
```

```python
X = boston.data
y = boston.target
x = range(0, boston.data.shape[0],1)
CRIMEN = 0
ZN = 1
INDUSTRIAL = 2
ROOMS = 5
DISTANCE = 7
```

```python
# GRAFICAMOS CARACTERISTICAS
plt.figure(figsize=(30,8))
plt.plot(x, X[:,INDUSTRIAL], "y-", label="Industrial")
plt.plot(x, X[:,DISTANCE], "b-", label="Distancia")
plt.plot(x, X[:,ROOMS], "c-", label="Habitaciones")
plt.rcParams.update({"font.size":18})
plt.legend()
plt.grid()
plt.show()
```

```python
# Obtenemos la media y la variancia de cada caracteristica
mean = []
variance = []
for i in range(boston.data.shape[1]):
    mean.append(statistics.mean(X[:,i]))
    variance.append(statistics.variance(X[:,i]))
    print(format(i) + " mean: {:.2f}".format(mean[i]) + ", variance: {:.2f}".format(variance[i]))
```

```python
fig, axs = plt.subplots(1,2,figsize=(7,4))

# Figuras en 3d
axs[1] = Axes3D(fig, elev=15, azim=30)
axs[1].scatter(X[:,DISTANCE], X[:,ROOMS], y, c="b", marker="o")
axs[1].set_xlabel("Distancia", fontsize=22)
axs[1].set_ylabel("Habitaciones", fontsize=22)
axs[1].set_zlabel("Boston", fontsize=22)
plt.show()
```
